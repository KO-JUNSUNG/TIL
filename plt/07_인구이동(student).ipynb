```python
import pandas as pd
```

```python
df = pd.read_csv('C:/Users/KOJUNSUNG/Desktop/lab/101_DT_1B26003_A01_20221207155740.csv',encoding='cp949') 
```

```python
# 맨위 5줄만 표시 하시오 
df=df.drop(['Unnamed: 56'],axis=1)
```

```python
df=df.fillna(0)
```

```python
# 전출지가 서울인 사람이 서울이 아닌곳으로 전입한 열을 선택하여 df2에 저장하세요 
# 서울(전출) -> 타지역(전입)
```

```python
df2=df.loc[(df['전출지별']=='서울특별시') & (df['전입지별']!='서울특별시')]
```

```python
df2.head()
```

```python
# 컬럼을 전출지별, 전입지별, 항목, 1970 년, 1971년, ..., 2022. 10월 로 정렬 하시오
```

```python
# 전출지별과 항목 컬럼을 삭제하고, index를 전입지 별로 지정하시오 
```

```python

df2.set_index('전입지별', inplace=True)
```

```python
df2=df2.drop(['전출지별','항목'], axis=1)
```

```python
df2
```

```python
# 서울에서 경기도로 이동한 인구 데이터를 df_su2ggd에 저장하세요. 
```

```python
df_su2ggd=df2[df2.index=='경기도'].iloc[0]
```

```python
df_su2ggd=df_su2ggd.drop('단위')
```

```python
df_su2ggd.astype(float)
```

```python
df_su2ggd.values
```

```python
df_su2ggd
```

```python
# 서울에서 경기도로 이동한 인구를 plot으로 나타내 보세요 
```

```python
fig=plt.figure(figsize=(20,10))
plt.plot(df_su2ggd)
plt.xticks(rotation=90)
plt.grid(True)
```
