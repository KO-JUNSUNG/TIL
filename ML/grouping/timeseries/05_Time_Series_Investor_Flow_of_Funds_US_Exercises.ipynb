### Step 7. index를 DatetimeIndex type으로 변경하세요

```python
df['newDate']=pd.to_datetime(df['Date']) # df.index=pd.to_datetime(df.index) 로 바로 들어가면 window linux time 기준으로(1970~) 정렬된다. --->string type 이기에 format 을 지정해주지 않으면 데이터이해를 못해서 윈도우타임을 가져오는 듯하다.
df.set_index('newDate',inplace=True)  
df.index #index 가 datetime index가 된 것을 확인했다.
df.head()
```

```python
df.drop('Date',axis=1)
```

### Step 8.  frequency 를 월 단위로 변경하고(월말 기준), 각 column의 월단위로 합산값을 구하세요

```python
df.resample('M').mean()
```

### Step 9. Step8에서의 결과중 NaN으로 채워진 값이 존재하는 row를 삭제 하세요

```python
df.resample('M').mean().dropna()
```
