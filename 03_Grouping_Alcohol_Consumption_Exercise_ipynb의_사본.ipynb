### Step 1. 필요한 라이브러리 들을 import 하세요

```python
import pandas as pd
import numpy as np
```

### Step 2. 다음 주소로 부터 dataset을 import 하세요 [address](https://raw.githubusercontent.com/myoh0623/dataset/main/drinks.csv). 

```python
drink=pd.read_csv("https://raw.githubusercontent.com/myoh0623/dataset/main/drinks.csv")
drink.head()
```

### Step 3. drinks 변수에 DataFrame을 할당 하세요

### Step 4. 각 continent별 beer_servings의 평균은 얼마인가요?

```python
drink.groupby('continent')['beer_servings'].mean()
```

### Step 5. 각 continent별 wine_servings의 통계량을 출력하세요

```python
drink.groupby('continent')['wine_servings'].describe()
```

### Step 6. 모든 column에 대해 continent별 각 음료간의 소비량의 평균을 구하세요

```python
drink.groupby('continent').mean()
```

### Step 7. 모든 column에 대해 continent별 각 음료간의 소비량의 중간값을 구하세요

```python
drink.groupby('continent').median()
```

### Step 8. spirit 소비량에 대한 평균, 최소, 최대 값을 구하세요

```python

drink.groupby('continent')['spirit_servings'].agg([np.mean,np.min,np.max]) #사용자함수를 적용하거나 여러 함수를 한번에 적용하려면 agg(aggregation) 함수를 사용하세요.
```
