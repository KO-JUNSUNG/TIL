```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
titanic=sns.load_dataset('titanic')
print(titanic.head())
```

```python
import matplotlib.pyplot as plt
```

```python
#style 설정
sns.set_style('darkgrid')
#그래프객체 생성
fig=plt.figure()
ax1=plt.subplot(121)
ax2=plt.subplot(122)

#그래프 그리기(선형회귀선: fit_reg=True)
sns.regplot(x='age', y='fare', data=titanic, ax=ax1)
#그래프 그리기(with no 선형회귀선)
sns.regplot(x='age', y='fare', data=titanic, ax=ax2, fit_reg=False)
plt.show()
```

```python
#커널밀도그래프/히스토그램 (distplot, kdeplot, histplo)
fig2=plt.figure(figsize=(15,5))
ax1=fig2.add_subplot(1,3,1)
ax2=fig2.add_subplot(1,3,2)
ax3=fig2.add_subplot(1,3,3)
#distplot(histogram 과 커널밀도함수(그래프와 x축 사이의 면적이 1이 되도록 그리는 밀도분포함수,cdf의 일종)를 동시에 그린다.)
sns.distplot(titanic['fare'], ax=ax1)
#kdeplot(커널밀도함수)
sns.kdeplot(titanic['fare'],ax=ax2)
#histogram(x와 x의 빈도를 나타내는 연속적인 그래프)
sns.histplot(x='fare',data=titanic,ax=ax3)
plt.show()
```

```python
#heatmap(상관관계분포도)
#pivot table로 범주형 변수를 각각 행, 열로 재구분하여 정리
table=titanic.pivot_table(index=['sex'],columns=['class'],aggfunc='size')

#히트맵그리기
sns.heatmap(table, #데이터프레임
            annot=True, fmt='d',#데이터값 표시 여부, 정수형 포맷
            cmap='YlGnBu', #컬러맵
            linewidths=.5, #구분선
            cbar=False)    #컬러바표시여부
plt.show()
```

```python
#strip plot and swarm plot 산점도, swarm plot= strip plot + 데이터의 분포형태를 보여준다.
"""seaborn.swarmplot(data=None, *, x=None, y=None, hue=None, order=None, hue_order=None, dodge=False, orient=None,
 color=None, palette=None, size=5, edgecolor='gray', linewidth=0, hue_norm=None, native_scale=False, formatter=None, 
 legend='auto', warn_thresh=0.05, ax=None, **kwargs)"""
fig3=plt.figure()
ax1=plt.subplot(1,2,1)
ax2=plt.subplot(1,2,2)

sns.stripplot(data=titanic, x='class', y='age',ax=ax1)
sns.swarmplot(data=titanic, x='class', y='age',ax=ax2)
plt.show()
```
