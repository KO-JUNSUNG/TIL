```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
titanic=sns.load_dataset('titanic')
print(titanic.head())
```

```python
import matplotlib.pyplot as plt
```

```python
#style 
sns.set_style('darkgrid')
# 
fig=plt.figure()
ax1=plt.subplot(121)
ax2=plt.subplot(122)

# (: fit_reg=True)
sns.regplot(x='age', y='fare', data=titanic, ax=ax1)
# (with no )
sns.regplot(x='age', y='fare', data=titanic, ax=ax2, fit_reg=False)
plt.show()
```

```python
#/ (distplot, kdeplot, histplo)
fig2=plt.figure(figsize=(15,5))
ax1=fig2.add_subplot(1,3,1)
ax2=fig2.add_subplot(1,3,2)
ax3=fig2.add_subplot(1,3,3)
#distplot(histogram  ( x   1   ,cdf )  .)
sns.distplot(titanic['fare'], ax=ax1)
#kdeplot()
sns.kdeplot(titanic['fare'],ax=ax2)
#histogram(x x    )
sns.histplot(x='fare',data=titanic,ax=ax3)
plt.show()
```

```python
#heatmap()
#pivot table    ,   
table=titanic.pivot_table(index=['sex'],columns=['class'],aggfunc='size')

#
sns.heatmap(table, #
            annot=True, fmt='d',#  ,  
            cmap='YlGnBu', #
            linewidths=.5, #
            cbar=False)    #
plt.show()
```

```python
#strip plot and swarm plot , swarm plot= strip plot +   .
"""seaborn.swarmplot(data=None, *, x=None, y=None, hue=None, order=None, hue_order=None, dodge=False, orient=None,
 color=None, palette=None, size=5, edgecolor='gray', linewidth=0, hue_norm=None, native_scale=False, formatter=None, 
 legend='auto', warn_thresh=0.05, ax=None, **kwargs)"""
fig3=plt.figure()
ax1=plt.subplot(1,2,1)
ax2=plt.subplot(1,2,2)

sns.stripplot(data=titanic, x='class', y='age',ax=ax1)
sns.swarmplot(data=titanic, x='class', y='age',ax=ax2)
plt.show()
```
