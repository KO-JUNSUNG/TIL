# solution

```python
import pandas as pd
df = pd.read_csv("../input/big-data-analytics-certification/t1-data1.csv")

df = pd.DataFrame(df.isnull().sum(), columns=['cnt_null'])
df = df['cnt_null'].sort_values(ascending=False)
print(df.index[0])
```
