```python
import numpy as np
import pandas as pd


train = pd.read_csv('/kaggle/input/big-data-analytics-certification-kr-2023-5th/train.csv')
test = pd.read_csv('/kaggle/input/big-data-analytics-certification-kr-2023-5th/test.csv')

print(train.head())
print(test.head())
```

```python
# 데이터 분할
y_train = train.pop('price')
x_train = train

print(x_train.head())
print(y_train.head())
```

```python
# 데이터 전처리
from sklearn.preprocessing import LabelEncoder

cols = ['model', 'transmission', 'fuelType']

for col in cols:
    encoder = LabelEncoder()
    x_train[col] = encoder.fit_transform(x_train[col])
    test[col] = encoder.fit_transform(test[col])
    
```

```python
# EDA
print(x_train.isna().sum())
print(test.isna().sum())
print(x_train.info(), test.info())
```

```python
import lightgbm as lgbm

model = lgbm.LGBMRegressor()
model.fit(x_train, y_train)
predicted_y_test = model.predict(test)
```

```python
submission = pd.read_csv('/kaggle/input/big-data-analytics-certification-kr-2023-5th/sample_submission.csv')
submission['price'] = predicted_y_test

print(submission.head())
```

```python
submission.to_csv('result.csv', index = False)
```
