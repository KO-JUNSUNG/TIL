구름이가 사는 구름 나라는 N개의 섬으로 이루어져 있다. 
각 섬에는 1부터 N까지의 번호가 붙어있고, 구름 나라는 사람들이 섬과 섬 사이를 편하게 이동할 수 있도록 다리를 M개 설치했다. 
설치된 다리들은 아래 특징을 만족한다.

- 모든 다리는 단방향으로만 이동할 수 있다.
- 서로 다른 두 섬을 잇는 다리는 최대 하나이다.
- 다리가 잇는 두 섬은 항상 다른 섬이다. 

구름이는 현재 K번 섬에 있고, 다른 섬들을 둘러본 뒤 다시 K번 섬으로 돌아오려고 한다. 구름이는 최소 경로를 택할 것이다. 구름이가 몇 번만에 돌아올 수 있는지 알아보자.
단, 구름이는 K번 섬 이외에 최소 하나 이상의 다른 섬을 방문해야 한다.
```python
예시:
    입력:
        5 6 4
        1 2
        1 4
        3 1
        4 2
        4 5
        5 3
    출력:
        4
```

````python
def shortest_path(graph, start, end):
    # 시작 정점에서부터 각 정점까지의 최단 거리를 저장하는 리스트
    distances = [float('inf')] * (len(graph) + 1)
    distances[start] = 0

    # 우선순위 큐를 사용하여 최소 거리를 관리
    queue = []
    heapq.heappush(queue, (0, start))

    while queue:
        current_distance, current_node = heapq.heappop(queue)

        # 이미 최단 거리를 구한 정점은 무시
        if current_distance > distances[current_node]:
            continue

        for adjacent_node, weight in graph[current_node]:
            distance = current_distance + weight

            # 현재 경로로 인접한 정점까지의 거리가 더 짧은 경우 갱신
            if distance < distances[adjacent_node]:
                distances[adjacent_node] = distance
                heapq.heappush(queue, (distance, adjacent_node))

    return distances[end]

# 입력 받기
N, M, K = map(int, input().split())

# 그래프 초기화
graph = [[] for _ in range(N + 1)]

# 다리 정보 입력 받기
for _ in range(M):
    a, b = map(int, input().split())
    graph[a].append((b, 1))  # 다리의 길이는 모두 1로 가정

# 최소 경로 찾기
min_distance = float('inf')
for i in range(1, N + 1):
    if i == K:
        continue
    distance1 = shortest_path(graph, K, i)  # K번 섬에서 i번 섬까지의 거리
    distance2 = shortest_path(graph, i, K)  # i번 섬에서 K번 섬까지의 거리
    if distance1 != float('inf') and distance2 != float('inf'):
        min_distance = min(min_distance, distance1 + distance2)

if min_distance == float('inf'):
    print(-1)  # 최소 경로가 없는 경우
else:
    print(min_distance)
````
