```python
from collections import deque

N, M = map(int, input().split())
Q = deque()

for i in range(M):
    op, cost = input().split()
    cost = int(cost)
    
    if op == "deposit":
        N = N + cost
        #   ,         .
        #         IndexError ,
        #       .
        while Q and Q[0] <= N:
            #     ,    .
            N = N - Q[0]
            Q.popleft()

    elif op == "pay":
        if N >= cost:
            N = N - cost

    elif op == "reservation":
        if not Q and N >= cost:
            N = N - cost
        else:
            # reservation   ,    .
            Q.append(cost)

print(N)
```
