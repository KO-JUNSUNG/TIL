- 정규식(Regular Expression)은 문자열에서 특정 패턴을 찾아내는 도구입니다. 패턴을 정의하고 이에 맞는 문자열을 찾아내는 작업을 수행할 수 있습니다.

- 정규표현식은 다양한 메타문자(Metacharacters)와 특수 문자(Special Characters)를 사용하여 표현됩니다. 이들을 조합하여 원하는 패턴을 구성할 수 있습니다. 아래는 사용된 정규표현식의 구성 요소와 의미에 대한 설명입니다:

- \d: 숫자에 매칭됩니다. \d+는 한 개 이상의 연속된 숫자에 매칭됩니다.
- |: OR 연산자로, 왼쪽 또는 오른쪽에 매칭되는 패턴 중 하나에 매칭됩니다.
- [-+*/]: 대괄호([]) 안에 있는 문자 중 하나에 매칭됩니다. 여기서는 -, +, *, / 중 하나에 매칭됩니다.

 re.findall(r'\d+|[-+*/]', expression)은 주어진 expression 문자열에서 숫자들과 -, +, *, / 중 하나에 매칭되는 패턴을 찾아내어 리스트로 반환합니다. 이렇게 패턴에 매칭된 문자열들을 토큰으로 분리할 수 있게 됩니다.

- 예를 들어, 입력으로 주어진 "10*3-10" 수식에서 위 정규식을 적용하면 ['10', '*', '3', '-', '10']와 같은 리스트를 얻게 됩니다. 이 리스트를 활용하여 수식을 분리하고, 연산을 수행할 수 있습니다.
