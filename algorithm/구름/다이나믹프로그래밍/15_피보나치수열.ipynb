```python
T = int(input())
# 6 -> 1,1,2,3,5,8
# 아하! 피보나치 수열의 6번째 수를 구하라는 문제구나!
# 문제를 해결하려면 피보나치수열을 구현하고(재귀로 해야할 것 같다) T 번째 수를 호출하면 되겠다.

def fivo(n:int)->int:
	if (n == 1) or (n == 2):
		return 1
	else:
		return fivo(n-1) + fivo(n-2)
	

print(fivo(T)%1000000007)
```

## 실제로 이렇게 풀었고, 테스트에서는 문제가 없었으나 제출에서는 시간 오버가 많이 발생했다. 왜일까?









## 그렇다, 문제는 __저장__ 이다!!

- 이런 방식을 하향식 방식이라고 부르며, 메모이제이션이라고 부른다. 

- 하위 문제에 대한 정답을 계산했는지 확인해가며 문제를 자연스럽게 재귀로 풀어나간다. 기존 재귀와 거의 동일하면서 이미 풀어봤는지 확인하며 재활용하는 효율적인 방식이다.




```python
# 해설 정답

import sys
sys.setrecursionlimit(111111)

# 이 정해 코드를 그냥 실행하면 스택 메모리 제한으로 인해 일부 테스트 케이스를 통과하지 못합니다.
# 아래 주석 처리된 코드와 같이 실행하면, 스택 메모리 제한을 임시적으로 조금 늘려서 문제를 해결할 수 있습니다.
# 시스템에 직접 접근하는 코드이므로, 일반적으로 코딩 테스트를 응시할 때는 사용하면 안되는 코드입니다!!

import resource
resource.setrlimit(resource.RLIMIT_STACK, (64 * 1024 * 1024, -1))


MOD = 10**9 + 7
F = [-1 for _ in range(100008)] # 0 ~ 100007 까지의 range 에 -1 을 나열했음. 즉 100008 개의 -1 을 만든 것. 

def fib(n):
    if F[n] != -1: # F 의 n 번째 원소가 -1 이 아니라면, F의 n번째 원소를 return 하라.
        return F[n]
    if n <= 2: # n이 2 이하라면 1을 return 하라. 
        return 1
    F[n] = (fib(n - 1) + fib(n - 2)) % MOD
    return F[n]
    
N = int(input())
print(fib(N))
```

```python
# 더 깔끔한 정답.
from collections import defaultdict

T = int(input())
dp = defaultdict(int)

def fivo(n:int)->int:
	if n <= 2:
		return 1
	if dp[n]:
		return dp[n]
	dp[n] = fivo(n-1) + fivo(n-2)
	return dp[n]   
	

print(fivo(T))
```
