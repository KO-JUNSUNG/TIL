```python
#  

import sys
sys.setrecursionlimit(111111)

#              .
#      ,          .
#    ,        !!

import resource
resource.setrlimit(resource.RLIMIT_STACK, (64 * 1024 * 1024, -1))


MOD = 10**9 + 7
F = [-1 for _ in range(100008)] # 0 ~ 100007  range  -1  .  100008  -1   . 

def fib(n):
    if F[n] != -1: # F  n   -1  , F n  return .
        return F[n]
    if n <= 2: # n 2  1 return . 
        return 1
    F[n] = (fib(n - 1) + fib(n - 2)) % MOD
    return F[n]
    
N = int(input())
print(fib(N))
```

```python
#   .
from collections import defaultdict

T = int(input())
dp = defaultdict(int)

def fivo(n:int)->int:
	if n <= 2:
		return 1
	if dp[n]:
		return dp[n]
	dp[n] = fivo(n-1) + fivo(n-2)
	return dp[n]   
	

print(fivo(T))
```
