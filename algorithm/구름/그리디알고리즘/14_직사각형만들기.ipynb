```python
# 나의 답
# 테스트케이스에서는 정답이었지만, 제출결과 굉장히 많은 런타임 에러가 발생했음.
# 반례를 생각해본 결과, 8,8,8,8 등의 입력에 대해서 set() 으로 처리를 해버리면 다리가 하나로 처리가 되므로 적절한 처리가 아님.

N = int(input())
S = list(map(int,input().split()))

# 4 개가 있으면 다 써야하고, 8개가 있으면 절반만 골라야 해. 직사각형을 만들 수 없다면 0 을 돌려줘야 해. 
S = sorted(S, reverse = True)
stick = []
for i in range(N): # 0,1,2,3,4,5,6,7
	if (i != (N-1) ) and S[i+1] == S[i]:
		stick.append(S[i])

stick = set(stick)
stick = sorted(stick,reverse=True)

ans = 0
for i in range(0,len(stick),2):
	ans += stick[i] * stick[i+1]
print(ans)
```

```python
# 해설 정답
import sys
input = sys.stdin.readline

N = int(input())
pair = []
cnt = [0 for _ in range(1000001)]
sticks = map(int, input().split())
for stick in sticks:
    cnt[stick] += 1

for length in range(1, 1000001):
    while cnt[length] > 1:
        cnt[length] -= 2
        pair.append(length)

pair.sort(reverse=True)
ans = 0
for i in range(1, len(pair), 2):
    ans += pair[i - 1] * pair[i]

print(ans)
```
