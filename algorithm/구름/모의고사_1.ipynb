```python
N,M = map(int,input().split())

from collections import defaultdict

dp = defaultdict(int)

for _ in range(M):
    temp = list(map(int,input().split()))[1:]
    for element in temp:
        if dp[element]:
            dp[element] += 1
        else:
            dp[element] = 1

print(dp.items())

# 제대로 카운트 되었다.
```

```python
ans = sorted(dp.items(),key = lambda x: (x[1],-x[0]) ,reverse=True)
max_value = max(item[1] for item in ans)
real_ans = [item[0] for item in ans if item[1] == max_value]
print(*real_ans)
```

```python
data = [(4, 4), (3, 4), (2, 4), (1, 4)]

# 뒤의 값들을 추출하여 집합으로 변환하고, 집합의 길이가 1이면 뒤의 값이 모두 같다는 의미이므로
# 앞의 자료를 출력, 그렇지 않으면 뒤의 값들 중 가장 큰 값을 출력
if len(set(item[1] for item in data)) == 1:
    keys = [item[0] for item in data]
else:
    keys = max(data, key=lambda x: x[1])[0]

print(keys)
```
