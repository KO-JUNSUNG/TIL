```python
N,M = map(int,input().split())

from collections import defaultdict

dp = defaultdict(int)

for _ in range(M):
    temp = list(map(int,input().split()))[1:]
    for element in temp:
        if dp[element]:
            dp[element] += 1
        else:
            dp[element] = 1

print(dp.items())

#   .
```

```python
ans = sorted(dp.items(),key = lambda x: (x[1],-x[0]) ,reverse=True)
max_value = max(item[1] for item in ans)
real_ans = [item[0] for item in ans if item[1] == max_value]
print(*real_ans)
```

```python
data = [(4, 4), (3, 4), (2, 4), (1, 4)]

#     ,   1     
#   ,         
if len(set(item[1] for item in data)) == 1:
    keys = [item[0] for item in data]
else:
    keys = max(data, key=lambda x: x[1])[0]

print(keys)
```

```python
data = [(4, 4), (3, 4), (2, 3), (1, 3)]

#     ,   1     
#   ,         
if len(set(item[1] for item in data)) == 1:
    keys = [item[0] for item in data]
else:
    max_value = max(item[1] for item in data)
    keys = [item[0] for item in data if item[1] == max_value]

print(keys)
```
