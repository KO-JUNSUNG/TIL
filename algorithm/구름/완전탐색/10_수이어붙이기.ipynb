```python
# 8788, 8887 두 경우의 수를 만들고 짝수번째 수가 짝수번째 수 +1 과 같을 경우 자릿수를 줄일 수 있다...
# permutation 을 쓸까?

import itertools 
import sys

# 입력
N = int(input())
A = list(map(int,input().split()))
A = sorted(A)

# 함수부

a = itertools.permutations(A,N)
ans = 1e18

for order in a:
	cur = order[0] # 가장 작은 수가 가장 앞에 나가야 가장 작은 수가 된다.
	# print(order)
	for i in range(1,N):
		if cur%10 == order[i]//10:  # 앞의 1의 자리수가 뒤의 10의 자리수와 같다면 합쳐버린다.
			cur = cur*10 + order[i]%10
		else:
			cur = cur*100 + order[i]
	# cur 을 바탕으로 다 합쳐진 다음 한번씩 비교
	ans = min(ans,cur)

print(ans)
```
