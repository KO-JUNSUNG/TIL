```python
# 8788, 8887         +1       ...
# permutation  ?

import itertools 
import sys

# 
N = int(input())
A = list(map(int,input().split()))
A = sorted(A)

# 

a = itertools.permutations(A,N)
ans = 1e18

for order in a:
	cur = order[0] #          .
	# print(order)
	for i in range(1,N):
		if cur%10 == order[i]//10:  #  1   10   .
			cur = cur*10 + order[i]%10
		else:
			cur = cur*100 + order[i]
	# cur       
	ans = min(ans,cur)

print(ans)
```
