```python
#             .

# 
N, M = map(int, input().split())
ants, aphids = [], []
for i in range(N):
    s = list(map(int, input().split()))
    for j in range(N):
        if s[j] == 1:
            ants.append([i, j])
        elif s[j] == 2:
            aphids.append([i, j])

# 2 = , 1 = , 0 = 
import sys
ans = len(ants)
for y1, x1 in ants:
    min_dist = sys.maxsize
    for y2, x2 in aphids:
        dist = abs(y1 - y2) + abs(x1 - x2)
        min_dist = min(min_dist, dist)
    if min_dist > M:
        ans -= 1

print(ans)

# distance = abs(a-c) + abs(b-d)
```
