```python
# 좌표를 뽑아내서 유클리드 거리를 구한 후 각 개미가 접근할 수 있는지 여부를 판별한다.

# 입력
N, M = map(int, input().split())
ants, aphids = [], []
for i in range(N):
    s = list(map(int, input().split()))
    for j in range(N):
        if s[j] == 1:
            ants.append([i, j])
        elif s[j] == 2:
            aphids.append([i, j])

# 2 = 진딧물, 1 = 개미, 0 = 비어있음
import sys
ans = len(ants)
for y1, x1 in ants:
    min_dist = sys.maxsize
    for y2, x2 in aphids:
        dist = abs(y1 - y2) + abs(x1 - x2)
        min_dist = min(min_dist, dist)
    if min_dist > M:
        ans -= 1

print(ans)

# distance = abs(a-c) + abs(b-d)
```
