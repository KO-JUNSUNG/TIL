# 두 이진트리를 병합하라. 중복되는 노드는 값을 합산한다.

- 입력

```python
트리 1


   1
  /\
 3   2
/
5

트리 2

2
/\
1 3
\  \
4   7
```


- 출력

```python
   3
  /\
 4 5
/\  \
5 4  7
```

````python
# 준비

class TreeNode:
    def __init__(self, val = 0, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right

def BuildTree(nodes, index):
    node = None
    # 현재 인덱스가 데이터 리스트의 범위를 벗어나지 않고, 노드가 None이 아닌 경우에만 노드를 생성합니다.
    if index < len(nodes) and nodes[index] is not None:
        # 현재 인덱스에 해당하는 값으로 새로운 노드를 생성합니다.
        node = TreeNode(nodes[index])
        # 왼쪽 자식 노드를 생성하기 위해 재귀적으로 build_tree 함수를 호출합니다.
        # 왼쪽 자식 노드의 인덱스는 현재 인덱스의 2배 + 1 입니다. 오른쪽은 2 *i + 2 입니다.
        node.left = BuildTree(nodes, 2*index + 1)
        node.right = BuildTree(nodes, 2 * index + 2)
    return node

tree_1 = [1,3,2,5,None,None,None]

tree_2 = [2,1,3,None,4,None,7]

rootnode_1 = BuildTree(tree_1,0)
rootnode_2 = BuildTree(tree_2,0)
````

````python
# 재귀탐색
# 너무 dfs나 bfs 에 매몰되고 있는 것은 아닌가 주의할 필요가 있다는 생각이 들었다.
class Solution:
    def BinaryTreeMerge(self, t1: TreeNode, t2: TreeNode) -> TreeNode:
        if t1 and t2:
            node = TreeNode(t1.val + t2.val)
            node.left = self.BinaryTreeMerge(t1.left, t2.left)
            node.right = self.BinaryTreeMerge(t1.right, t2.right)
        
            return node
        else:
            return t1 or t2
````

````python

# BFS 방법
from collections import deque

class Solution:
    def mergeTrees(self, t1, t2):
        if not t1:
            return t2
        if not t2:
            return t1
        
        queue = deque([(t1, t2)])
        
        while queue:
            node1, node2 = queue.popleft()
            
            node1.val += node2.val
            
            if node1.left and node2.left:
                queue.append((node1.left, node2.left))
            elif not node1.left:
                node1.left = node2.left
            
            if node1.right and node2.right:
                queue.append((node1.right, node2.right))
            elif not node1.right:
                node1.right = node2.right
        
        return t1
````

````python
# DFS 방법

class Solution:
    def mergeTrees(self,t1,t2):
        # 아래의 코드 조각은 t1과 t2 중 하나라도 존재하지 않을 경우, 다른 존재하는 트리를 반환하는 부분입니다. 이 부분은 재귀적인 방식으로 트리를 병합할 때 종료 조건을 설정하기 위해 사용됩니다.
        if not t1:
            return t2
        if not t2:
            return t1
        
        t1.val += t2.val
        t1.left = self.mergeTrees(t1.left, t2.left)
        t1.right = self.mergeTrees(t1.right, t2.right)

        return t1
    
````
