````python
T = "abcabcbb"

def subStringLength(s:str)->int:
    # 입력을 생각하고, 필요한 출력을 생각하고, 거기에 필요한 알고리즘을 생각해.
    used = {}
    max_length = start = 0
    for index, char in enumerate(s):
        # 이미 등장했던 문자라면, 'start' 위치 갱신  == 중복문자 판별 및 현재 슬라이딩 윈도우 바깥의 문자 무시
        if char in used and start <= used[char]:
            start = used[char] + 1 # 현재 위치 + 1
        else: # 최대 부분 문자열 길이 갱신
            max_length = max(max_length, index - start + 1)
        # 현재 문자를 dict 에 삽입
        used[char] = index
    return max_length
````

- 좌측 포인터는 start, 우측 포인터는 index 다. 단 두 포인터 모두 좌측에서 출발한다.
- 두번쨰 포인터는 다음과 같이 계속 오른쪽으로 확장한다.

```python

```

- 이미 등장한 문자라면 used 에 있을 것이고 이 경우 첫번쨰 포인터인 start 를 used[char] + 1 로 이동한다.
