어느 고등학교에서 실시한 1000명의 수학 성적을 토대로 통계 자료를 만들려고 한다.

이때, 이 학교에서는 최빈수를 이용하여 학생들의 평균 수준을 짐작하는데, 여기서 최빈수는 특정 자료에서 가장 여러 번 나타나는 값을 의미한다.

다음과 같은 수 분포가 있으면,

10, 8, 7, 2, 2, 4, 8, 8, 8, 9, 5, 5, 3

최빈수는 8이 된다.

최빈수를 출력하는 프로그램을 작성하여라 (단, 최빈수가 여러 개 일 때에는 가장 큰 점수를 출력하라).

[제약 사항]

학생의 수는 1000명이며, 각 학생의 점수는 0점 이상 100점 이하의 값이다.
 
[입력]

첫 번째 줄에 테스트 케이스의 수 T가 주어진다.

각 테스트 케이스의 첫 줄에는 테스트 케이스의 번호가 주어지고 그 다음 줄부터는 점수가 주어진다.

[출력]

#부호와 함께 테스트 케이스의 번호를 출력하고, 공백 문자 후 테스트 케이스에 대한 답을 출력한다.

```python
# numpy 방법

import numpy as np
 
def find_mode(numbers):
    modes, counts = np.unique(numbers, return_counts=True)
    max_count = np.max(counts)
    mode_values = modes[counts == max_count]
    max_mode = np.max(mode_values)
    return max_mode
 
 
T = int(input())
for i in range(1, T + 1):
    N = int(input())
    M = np.array(list(map(int, input().split())))
    ans = find_mode(M)
    print(f'{N} {int(ans)}')
```

```python
# dict 사용 최빈수 구하기 방법
'''
my_dict = {'key1': 'value1', 'key2': 'value2', 'key3': 'value3'}
# 단일 키를 사용하여 값을 가져옴
value1 = my_dict['key1']
print(value1)  # 출력: value1
 
# get() 메서드를 사용하여 값을 가져옴
value2 = my_dict.get('key2')
print(value2)  # 출력: value2
'''
 

def find_mode(arr):
    counts = {}
    #  counts라는 딕셔너리를 활용하여 각 숫자의 빈도수를 저장
    # 입력리스트를 순회하면서 각 숫자의 빈도수를 업데이트
    for i in arr:
        counts[i] = counts.get(i, 0) + 1  
        # counts.get(num, 0)는 딕셔너리 counts에서 num을 키로 가진 값이 존재하지 않을 경우 기본값 0을 반환.
        # get() 메서드는 딕셔너리에서 특정 키의 값을 가져오는데, 키가 존재하지 않는 경우 기본값으로 0을 반환합니다. 따라서 counts.get(num, 0)은 num이 counts 딕셔너리에 존재한다면 해당 키의 값을 반환하고, 존재하지 않는다면 0을 반환.
        # 예를 들어 counts[41] = counts.get(41, 0) + 1 -> counts.get(41, 0) = 0, 왜냐하면 처음 시행 때 집합은 공집합이니까.
        # count[41] 은 key 가 41인 값을 가져오라는 명령
        # dict 는 key 를 이용하여 indexing 이 가능.
 
         
 
    max_count = max(counts.values())
    mode_values = [i for i, count in counts.items() if count == max_count]
    max_mode = max(mode_values)
    return max_mode
 
 
T = int(input())
for i in range(1, T + 1):
    N = int(input())
    M = list(map(int, input().split()))
    ans = find_mode(M)
    print(f'#{N} {int(ans)}')
```

```python
# 0~ 100 점까지의 0으로 가득 찬 리스트를 만들고, 카운트 될 때 마다 1씩 더하는 방법

     
for i in range(1, int(input()) + 1):
    _ = input()  # 테스트케이스 번호
    grades = list(map(int, input().split()))  #점수
    freq = [0] * 101  #0~100점까지의 빈도를 구하기 위함
    mode = 0  #최빈값
    for grade in grades:
        freq[grade] += 1 #현재점수의 빈도상승
        if freq[grade] >= freq[mode]: mode = grade #현재점수 빈도가가 최빈값 이상이면 최빈수 변경
    print(f"#{i} {mode}")
```
